```python
# Implement weather modeling using the quadratic solution with hard coding variables (fixed variable values give in program)

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(a, b, c, x):
    y = a * x**2 + b * x + c
    return y

def plot_temperature_over_time(a, b, c, time_range):
    time = np.array(time_range)
    temperature = quadratic_solution(a, b, c, time)

    plt.plot(time, temperature)
    plt.title("Temperature Over Time")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.grid(True)
    plt.show()

def main():
    print("Weather Modeling using Quadratic Solution")

    a_hardcoded, b_hardcoded, c_hardcoded = 2, -10, 0
    time_range = np.linspace(0, 10, 100)

    plot_temperature_over_time(a_hardcoded, b_hardcoded, c_hardcoded, time_range)

if __name__ == "__main__":
    main()
```

```python
# Implement weather modeling using the quadratic solution with hard coding variables for multiple sets of inputs
# (multiple sets of variable values given in program or file or user input and print all sets of graphs in one plot and specify those sets values)


import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time,a,b,c):
  temperature=a*(time**2)+b*time+c
  return temperature

coefficients_list={(0.5, 2.3, 10), (0.025, 1.2, 18), (1.7, 0.8, 12)}

time_values=np.arange(0,51,1)

for i, (a,b,c) in enumerate(coefficients_list):
  temperature_values=quadratic_temperature_model(time_values,a,b,c)
  label=f'Set(i+1): a={a}, b={b}, c={c}'
  plt.plot(time_values,temperature_values,label=label)

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic temp model with multiple sets of coefficets')
plt.legend()
plt.grid(True)
plt.show()
```
